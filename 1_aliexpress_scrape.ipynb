# Scraping AliExpress Ecommerce Site: DSE HACKATHON

## Import libraries

```python
import requests as req
from bs4 import BeautifulSoup
import pandas as pd
```

## Define function to fetch website html data

```python
def fetch_html_data(web_address):
    try:
        print(f"Fetching data from {web_address}...")
        res = req.get(web_address)
        return res
    except req.exceptions.RequestException as e:
        print('Stopped:', e)
    except TypeError as e:
        print('Stopped:', e)
```

# Define function to convert html data to BeautifulSoup object 

```python
def convert_web_data_to_beautiful_soup_obj(web_data):
    try:
        print("\nCreating BeautifulSoup object...")
        soup_obj = BeautifulSoup(web_data.text, "html.parser")
        print("Success! Object created!")
        return soup_obj
    except Exception as e:
        print("Stopped:", e)
```

### Declare website source urls

```python
webroot_fetch_url = "https://www.aliexpress.com/"
home_and_garden_url = "https://www.aliexpress.com/p/calp-plus/index.html?spm=a2g0o.categorymp.allcategoriespc.1.4bb8VpsPVpsPNp&categoryTab=home_%2526_garden"
```

### Fetch root url web data & convert to BeautifulSoup Object

```python
response = fetch_html_data(webroot_fetch_url)
soup = convert_web_data_to_beautiful_soup_obj(response)
```

```python
print(soup.prettify())
```

### Fetch home&garden category web data & convert to BeautifulSoup object

```python
response = fetch_html_data(home_and_garden_url)
soup = convert_web_data_to_beautiful_soup_obj(response)
```

#### Print home & garden category soup object

```python
print(soup.prettify())
```

### Selecting all home & garden item names

```python
all_home_and_garden_items = soup.find_all("h1", class_="nXeOv")
print(all_home_and_garden_items)
```
